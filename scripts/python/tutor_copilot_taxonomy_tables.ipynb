```python
df_output = pd.read_csv("/content/drive/Shareddrives/NSSA Research/FEV/FEV_AICopilot/data/replication/ana/tutor_copilot_analysis/datafiles/messages/annotated_huggingface/strategy_classified.csv")

df_output.info()
```

```python
df_sessions = pd.read_csv("/content/drive/Shareddrives/NSSA Research/FEV/FEV_AICopilot/data/replication/ana/tutor_copilot_analysis/datafiles/filtered_copilot_data.csv", usecols=['SESSION_ID', 'TUTOR_COPILOT_ASSIGNMENT'])
df_sessions.columns = df_sessions.columns.str.lower()

df_sessions.info()
```

```python
df_filtered = pd.merge(df_output, df_sessions, on='session_id', how='inner')
df_filtered.info()
```

```python
df_testset = pd.read_csv("/content/drive/Shareddrives/NSSA Research/FEV/FEV_AICopilot/data/datafiles/annotated/test.csv")

df_testset.info()
```

```python
for col in df_testset.columns:
    if df_testset[col].dtype == 'object':
        df_testset[col] = df_testset[col].str.lower()

display(df_testset.head())
```

```python
common_cols = ['context', 'response', 'session_id']
merged_df = pd.merge(df_filtered, df_testset, on=common_cols, how='inner')

display(merged_df.head())
display(merged_df.info())
```

```python
strategies_cols = [col for col in merged_df.columns if col.startswith('strategy_') & col.endswith('_class')]

percentage_of_positive_cases = {}
for col in strategies_cols:
    percentage = (merged_df[col].sum() / len(merged_df)) * 100
    percentage_of_positive_cases[col] = percentage

percentage_merged_df = pd.DataFrame.from_dict(percentage_of_positive_cases, orient='index', columns=['Percentage of Positive Cases'])

display(percentage_merged_df)
```

```python
CLASSIFIER_rename = {
    "strategies-2": "strategy_askquestion_class",
    "strategies-4": "strategy_solutionstrategy_class",
    "strategies-5": "strategy_promptexplanation_class",
    "strategies-6": "strategy_encouragestudent_class",
    "strategies-7": "strategy_affirmcorrect_class",
    "strategies-8": "strategy_answerexplanation_class",
    "strategies-9": "strategy_promptretry_class",
}
```

```python
from sklearn.metrics import f1_score
import numpy as np

# Reverse the CLASSIFIER_rename dictionary to map strategy number to column name
strategy_number_to_col = {int(k.split('-')[1]): v for k, v in CLASSIFIER_rename.items()}

f1_scores = {}
for strategy_col in strategies_cols:
    # Get the corresponding strategy number from the reversed dictionary
    strategy_number = None
    for num, col_name in strategy_number_to_col.items():
        if col_name == strategy_col:
            strategy_number = num
            break

    if strategy_number is not None:
        # Create binary labels
        y_true_binary = (merged_df['strategies_response'] == strategy_number).astype(int)
        y_pred_binary = merged_df[strategy_col].fillna(0).astype(int)

        # Calculate F1 score
        # Handle cases where there are no positive true labels
        if np.sum(y_true_binary) == 0:
            f1 = np.nan # Or 0.0 depending on how you want to represent this
        else:
            f1 = f1_score(y_true_binary, y_pred_binary, average='binary')
        f1_scores[strategy_col] = f1
    else:
        f1_scores[strategy_col] = np.nan # Handle cases where no mapping is found


f1_scores_df = pd.DataFrame.from_dict(f1_scores, orient='index', columns=['F1 Score'])

display(f1_scores_df)
```

```python
CLASSIFIER_STRATEGY_NL = {
    "strategy_askquestion_class": "Ask Question to Guide Thinking",
    "strategy_solutionstrategy_class": "Give Solution Strategy",
    "strategy_promptexplanation_class": "Prompt Student to Explain",
    "strategy_encouragestudent_class": "Encourage Student in Generic Way",
    "strategy_affirmcorrect_class": "Affirm Student's Correct Attempt",
    "strategy_answerexplanation_class": "Give Away Answer/Explanation",
    "strategy_promptretry_class": "Ask Student to Retry",
}
```

```python
combined_df = percentage_merged_df.merge(f1_scores_df, left_index=True, right_index=True)

# Replace the index with the descriptive labels
combined_df = combined_df.rename(index=CLASSIFIER_STRATEGY_NL)

display(combined_df)
```

```python
strategies_cols = [col for col in df_filtered.columns if col.startswith('strategy_') & col.endswith('_class')]

percentage_of_positive_cases = {}
for col in strategies_cols:
    percentage = (df_filtered[col].sum() / len(df_filtered)) * 100
    percentage_of_positive_cases[col] = percentage

percentage_filtered_df = pd.DataFrame.from_dict(percentage_of_positive_cases, orient='index', columns=['Percentage of Positive Cases'])

display(percentage_filtered_df)
```

```python
combined_df = percentage_filtered_df.merge(f1_scores_df, left_index=True, right_index=True)

# Replace the index with the descriptive labels
combined_df = combined_df.rename(index=CLASSIFIER_STRATEGY_NL)

display(combined_df)
```

```python
df_moments = pd.read_csv("/content/drive/Shareddrives/NSSA Research//FEV/FEV_AICopilot/data/replication/ana/tutor_copilot_analysis/datafiles/messages/annotated_huggingface/moments_classified.csv")

df_moments.info()
```

```python
df_filtered = pd.merge(df_moments, df_sessions, on='session_id', how='inner')
df_filtered.info()
```

```python
common_cols = ['context', 'prev_response', 'session_id']

merged_df_moments = pd.merge(df_moments, df_testset, on=common_cols, how='inner')

display(merged_df_moments.head())
display(merged_df_moments.info())
```

```python
moments_cols = [col for col in merged_df.columns if col.startswith('moments_') & col.endswith('_pred')]

percentage_of_positive_occurrences = {}
for col in moments_cols:
    percentage = (merged_df[col].sum() / len(merged_df)) * 100
    percentage_of_positive_occurrences[col] = percentage

percentage_moments_df = pd.DataFrame.from_dict(percentage_of_positive_occurrences, orient='index', columns=['Percentage of Positive Occurrences'])

display(percentage_moments_df)
```

```python
f1_scores = {}
for col in moments_cols:
    moments_number = int(col.split('_')[1])

    # Create binary labels
    y_true_binary = (merged_df['events_response'] == moments_number).astype(int)
    y_pred_binary = merged_df[col].fillna(0).astype(int)

    # Calculate F1 score
    f1 = f1_score(y_true_binary, y_pred_binary, average='binary')
    f1_scores[col] = f1

f1_scores_df = pd.DataFrame.from_dict(f1_scores, orient='index', columns=['F1 Score'])

display(f1_scores_df)
```

```python
combined_moments_df = percentage_moments_df.merge(f1_scores_df, left_index=True, right_index=True)

# Replace the index with the descriptive labels
combined_moments_df = combined_moments_df.rename(index=MOMENT_ID_2_NAME)

display(combined_moments_df)
```
